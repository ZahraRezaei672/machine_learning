```python
from PIL import Image
```

```python
img_path = './dataset/leo.jpg'
image = Image.open(img_path)
```

```python
print('Format:', image.format)
print('Mode:', image.mode)
print('Size:', image.size)
```

```python
import matplotlib.pyplot as plt
plt.imshow(image)
plt.show()
```

```python
image_resized = image.resize((175, 175))
plt.imshow(image_resized)
plt.show()
```

```python
image_transpose_top_bottom = image.transpose(Image.FLIP_TOP_BOTTOM)
plt.imshow(image_transpose_top_bottom)
plt.show()
```

```python
image_transpose_left_right = image.transpose(Image.FLIP_LEFT_RIGHT)
plt.imshow(image_transpose_left_right)
plt.show()
```

```python
image_rotation = image.rotate(45)
plt.imshow(image_rotation)
plt.show()
```

```python
image_rotation2 = image.rotate(-45)
plt.imshow(image_rotation2)
plt.show()
```

```python
image_rotation2.save('leo_rotate.jpg', format='JPEG')
```
