```python
import skimage.io as io

import matplotlib.pyplot as plt
import numpy as np

img = io.imread('./dataset/leo.jpg')
plt.imshow(img)
plt.show()
```

```python
img.shape
```

```python
red = img[:, :, 0]
green = img[:, :, 1]
blue = img[:, :, 2]
```

```python
plt.imshow(red, cmap='Reds')
plt.show()
```

```python
plt.imshow(green, cmap='Greens')
plt.show()
```

```python
plt.imshow(blue, cmap='Blues')
plt.show()
```

```python
from PIL import Image

img2 = Image.open('dataset/leo.jpg').convert('L')
img2.save('./dataset/grayleo.jpg')
```
